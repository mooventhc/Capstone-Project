# Feauture Selection-Using statistical test

Chi Square Test
           *Chi Square test is works when when the independent and dependent variable is categorical.
           *It is a non parametric test deals with degree of freedom and alpha.
           *Frequency should be greater than 5.

Checking whether revenue is dependent on month

h0:Month is independent on revenue
h1:Month is dependent on revenue

```python
a=pd.crosstab(data["Month"],data["Revenue"])
a
```

```python
chi_crit=st.chi2.isf(0.05,9)
chi_crit
```

```python
overall=st.chi2_contingency(a)
overall
```

Inference:
Here staistical value is greater than critical value.so we reject h0.Thus month is dependent on revenue.

```python
import statsmodels.api         as     sm
from   statsmodels.formula.api import ols
from   statsmodels.stats.anova import anova_lm

model = ols('Revenue ~ Administrative_Duration + Administrative',data = data).fit()
anova_table=anova_lm(model)
anova_table
```

```python
# P_value is less than 0.05 ; 
# Hence we reject the null hypothesis  
# There is a Significant effect due to Administrative_Duration and Administrative
```

```python
model = ols('Revenue ~ Informational + Informational_Duration',data = data).fit()
anova_table=anova_lm(model)
anova_table
```

```python
# Informational          - P_value is less than 0.05- Hence we reject the null hypothesis- there is Significant effect
# Informational_Duration - P_value is less than 0.05- Hence we accept the null hypothesis- there is Significant effect
```

```python
#Backward Elimination
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X,y,test_size=0.3, random_state = 0,stratify= y)

# checking the shapes

print("Shape of x_train :", X_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", X_test.shape)
print("Shape of y_test :", y_test.shape)
```

```python
# MODELLING

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# evaluating the model
print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score(X_test, y_test))

# confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.rcParams['figure.figsize'] = (6, 6)
sns.heatmap(cm ,annot = True)

# classification report
cr = classification_report(y_test, y_pred)
print(cr)
```

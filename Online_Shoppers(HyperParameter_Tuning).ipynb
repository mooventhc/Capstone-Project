```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/sharmaroshan/Online-Shoppers-Purchasing-Intention/master/online_shoppers_intention.csv')
data.head()
```

```python
data[['Revenue',"Weekend","TrafficType","Region","Browser","OperatingSystems","SpecialDay"]]=data[['Revenue',"Weekend","TrafficType","Region","Browser","OperatingSystems","SpecialDay"]].astype(str)
data
```

```python
data['Revenue']=data['Revenue'].map({'True':1,'False':0}).astype(int)
data
```

```python
cs = [i for i in data.columns if data[i].dtype == 'object']
df = pd.get_dummies(data,columns =cs,drop_first=True)
```

```python
df
```

```python
X=df.drop('Revenue',1)
y=df['Revenue']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import statsmodels.api as sm
X_1 = sm.add_constant(X)
model = sm.OLS(y,X_1).fit()
model.pvalues
```

```python
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

```python
selected_features_BE
```

```python
selected_features_BE.append("Revenue")
```

```python
df= df[selected_features_BE]
df
```

```python
df.info()
```

```python
X=df.drop(['Revenue'],axis=1)
y=df['Revenue']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X,y,test_size=0.3, random_state = 0,stratify= y)

# checking the shapes

print("Shape of x_train :", X_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", X_test.shape)
print("Shape of y_test :", y_test.shape)
```

### Analyse of the performance with the help of confusion Matrix

```python
def tn(y_true, y_pred): return confusion_matrix(y_true, y_pred)[0, 0]
def fp(y_true, y_pred): return confusion_matrix(y_true, y_pred)[0, 1]
def fn(y_true, y_pred): return confusion_matrix(y_true, y_pred)[1, 0]
def tp(y_true, y_pred): return confusion_matrix(y_true, y_pred)[1, 1]
```

```python
def precision(y_true,y_pred): return tp(y_true, y_pred)/(tp(y_true, y_pred)+fp(y_true, y_pred))
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# evaluating the model
print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score(X_test, y_test))

# confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# classification report
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
print("Logistic Regression True Negative",tn(y_test, y_pred))
print("Logistic Regression False Positive",fp(y_test, y_pred))
print("Logistic Regression False Negative",fn(y_test, y_pred))
print("Logistic Regression True Positive",tp(y_test, y_pred))
print("Precision",precision(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
dual=[True,False]
max_iter=[100,110,120,130,140]
C = [1.0,1.5,2.0,2.5]

param_grid = dict(dual=dual,max_iter=max_iter,C=C)

import time

lr = LogisticRegression(penalty='l2')
grid = GridSearchCV(estimator=lr, param_grid=param_grid, cv = 3, n_jobs=-1)

start_time = time.time()
grid_result = grid.fit(X, y)
# Summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
print("Execution time: " + str((time.time() - start_time)) + ' ms')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

lgt=LogisticRegression(**grid_result.best_params_)

lgt.fit(X_train, y_train)

y_pred = lgt.predict(X_test)

# evaluating the model
print("Training Accuracy :", lgt.score(X_train, y_train))
print("Testing Accuracy :", lgt.score(X_test, y_test))

# confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# classification report
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
from sklearn.model_selection import RandomizedSearchCV
random = RandomizedSearchCV(estimator=lr, param_distributions=param_grid, cv = 3, n_jobs=-1)

start_time = time.time()
random_result = random.fit(X, y)
# Summarize results
print("Best: %f using %s" % (random_result.best_score_, random_result.best_params_))
print("Execution time: " + str((time.time() - start_time)) + ' ms')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

lgt=LogisticRegression(**random_result.best_params_)

lgt.fit(X_train, y_train)

y_pred = lgt.predict(X_test)

# evaluating the model
print("Training Accuracy :", lgt.score(X_train, y_train))
print("Testing Accuracy :", lgt.score(X_test, y_test))

# confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# classification report
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
print("Logistic Regression True Negative",tn(y_test, y_pred))
print("Logistic Regression False Positive",fp(y_test, y_pred))
print("Logistic Regression False Negative",fn(y_test, y_pred))
print("Logistic Regression True Positive",tp(y_test, y_pred))
```
